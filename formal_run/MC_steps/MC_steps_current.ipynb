Check if the current converge amongs different MC steps

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("/home/ggu7596/project/optimal_control/motorsim_double_track/toy_model_v2/post_analysis/analysis_classes")
from dynamics_analysis import nonequilibrium_analysis
```

```python
noneq = nonequilibrium_analysis()
MC_steps_dirs = ["MC_steps_1","MC_steps_10","MC_steps_100"]
h_dirs = ["h_1.15","h_2.25","h_3.15"]
path = "/home/ggu7596/project/optimal_control/motorsim_double_track/toy_model_v2/formal_run/MC_steps"
current = np.zeros((len(MC_steps_dirs),len(h_dirs),2))
for i,MC_steps_dir in enumerate(MC_steps_dirs):
    for j,h_dir in enumerate(h_dirs):
        sub_path = path+"/"+MC_steps_dir+"/"+h_dir
        current[i,j,0], current[i,j,1] = noneq.calculate_current(sub_path)
```

```python
for i,MC_steps_dir in enumerate(MC_steps_dirs):
    plt.errorbar(np.arange(3),current[i,:,0], yerr = current[i,:,1],label = "MC steps = "+str(MC_steps_dir.split("_")[-1]))
plt.legend()
plt.xticks(np.arange(3),[float(h_dir.split("_")[-1])*2 for h_dir in h_dirs])
plt.ylim(0,0.0003)
plt.xlabel("h")
plt.ylabel("current")
plt.show()
    
```
